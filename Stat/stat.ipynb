```python
def getTimes(fileName):
    file = open(fileName)
    lines = list(file.read().splitlines())

    timesAstar = list()
    timesSecondAlgo = list()

    cnt = 1
    for line in lines:
        if cnt == 7:
            timesAstar.append(float(line[5:]))
        if cnt == 13:
            timesSecondAlgo.append(float(line[5:]))
            cnt = 0
        cnt += 1

    return [timesAstar, timesSecondAlgo]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def printGraph(fileName, secondAlgo):
    timesAstar, timesSecondAlgo = getTimes(fileName)
    print("Общее время работы A*: \t\t", sum(timesAstar))
    print("Общее время работы "+ secondAlgo +": \t", sum(timesSecondAlgo))
    lst = []
    for i in range(len(timesAstar)):
        lst.append(i + 1)

    plt.figure(figsize=(18, 9))
    plt.semilogy(lst, timesAstar, color='blue', label='A*')
    plt.semilogy(lst, timesSecondAlgo, color='red', label=secondAlgo)
    plt.title('A* и ' + secondAlgo)
    plt.ylabel('Время, в секундах')
    plt.xlabel('Номер итерации')
    plt.legend()
```

```python
printGraph("dataLPAStar", "LPA*")
```

```python
printGraph("dataLPAStar_short", "LPA*")
```

```python
printGraph("dataDStarLite_short", "D* Lite")
```

```python
printGraph("dataDStarLite", "D* Lite")
```
